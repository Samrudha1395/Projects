```python
import random
```

```python
class Quiz:
     def __init__(self,question,answer):
            self.question = question
            self.answer = answer
         
            
            
question_prompts =open ("quiz_questions.txt", "r") 
quiz = question_prompts.read()


questions = [
    Quiz(quiz[1:93],"a"),
    Quiz(quiz[93:209],"a"),
    Quiz(quiz[209:300],"a"),
    Quiz(quiz[300:400],"b"),
    Quiz(quiz[400:493],"b"),
    Quiz(quiz[493:603],"b"),
    Quiz(quiz[603:701],"c"),
    Quiz(quiz[701:803],"c"),
    Quiz(quiz[803:902],"c"),
    Quiz(quiz[902:1002],"c"), 
]    

random.shuffle(questions)

             

def start_quiz(questions):
    points=0
    for question in questions:
        answer= input(question.question)
        if answer == question.answer:
             points += 1
    if points<=3:
        print("You got", points, "out of", len(questions),". Your grade is C. Try harder.")
    elif points>3 and points <=8:
        print("You got", points, "out of", len(questions),". Your grade is B. Keep going!")
    else:
        print("You got", points, "out of", len(questions),". Your grade is A. Well done!")
        
start_quiz(questions)    

```
